```python
text_file = open('/content/File1.txt', 'r')
```

```python
text=text_file.read()
```

```python
print(text)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
  
tokenizer = AutoTokenizer.from_pretrained("tuner007/pegasus_paraphrase")

model = AutoModelForSeq2SeqLM.from_pretrained("tuner007/pegasus_paraphrase")
```

```python
sentences = text.split('.')
l=''
i=0
for sentence in sentences:
  tokens = tokenizer(sentence, truncation=True, padding="longest", return_tensors="pt")
  summary = model.generate(**tokens)
  a=tokenizer.decode(summary[0])
  l=l+a
  i+=1
  print(i)

  


```

```python
l
```

```python
p=l.replace("<pad>","")
```

```python
q=p.replace("</s>","")
```

```python
q
```

```python
text_file1 = open("/content/output.txt", "w")

text_file1.write(q)
text_file1.close()
```
